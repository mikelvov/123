```python
#Приводим категорию origin к числу
```

```python
data = pd.get_dummies(data, columns=["origin"], drop_first=True)
```

```python
#Нормализуем model year и cylinders
```

```python
a = data['model year'].mean()
v = data['model year'].std()
data['model year'] = (data['model year'] - a)/v
```

```python
a1 = data['cylinders'].mean()
v1 = data['cylinders'].std()
data['cylinders'] = (data['cylinders'] - a1)/v1
```
