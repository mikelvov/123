### Заменим пропуски средним по весовой категории

```python
data.groupby(data['model year']).horsepower.mean()
```

```python
data['weight_category'] = data.weight.apply(lambda w: int(w / 500))
```

```python
data.groupby(data.weight_category).horsepower.mean()
```

```python
data['horsepower_nona'] = data.groupby(data.weight_category).horsepower.apply(lambda h: h.fillna(h.mean()))
```

```python
data[data.horsepower.isna()]
```

```python
data.horsepower = data.horsepower_nona
```

```python
data.drop('horsepower_nona', axis=1, inplace=True)
data.drop('weight_category', axis=1, inplace=True)
```

```python
data.horsepower.isna().sum()
```

```python
data.horsepower[32]
```

### Приводим категории к числу

```python
data = pd.get_dummies(data, columns=['origin','model year'], drop_first=True)
```

<font color="red"> Лучше использовать стандартные нормалайзеры типа https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html </font>

```python
data['cylinders'] = (data['cylinders'] - data['cylinders'].mean())/data['cylinders'].std()
data['displacement'] = (data['displacement'] - data['displacement'].mean())/data['displacement'].std()
data['horsepower'] = (data['horsepower'] - data['horsepower'].mean())/data['horsepower'].std()
data['weight'] = (data['weight'] - data['weight'].mean())/data['weight'].std()
data['acceleration'] = (data['acceleration'] - data['acceleration'].mean())/data['acceleration'].std()
```

### Train/Test

```python
from sklearn.model_selection import train_test_split

X = data.drop('mpg', axis=1).to_numpy()
y = data.mpg.to_numpy()


print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

### SGDR Regression


```python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor(penalty='l1', alpha=0.05, max_iter=10000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```

### Grid Search, определим лучшие коэффициенты


```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty': ['l1', 'l2'], 'alpha': [0.01, 0.05, 0.1, 0.5, 1], 'max_iter': [1000, 5000, 10000]}
```

```python
model = SGDRegressor()
```

```python
gs_model = GridSearchCV(model, parameters, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
```

```python
gs_model.fit(X_train, y_train)
```

```python
gs_model.best_params_
```

### Расчет с лучшими коэффициентами

```python
model = SGDRegressor(alpha=0.01, max_iter=5000, penalty='l1')
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
```
